```python
import torch
import matplotlib.pyplot as plt
import numpy as np

# Configurar para reprodutibilidade
torch.manual_seed(42)

# 1. Calcular derivada de y = 3x² + 2x + 1 em x = 2
print("=== Exercício 1A: Derivada Simples ===")
x = torch.tensor(2.0, requires_grad=True)
y = 3 * x**2 + 2 * x + 1
print(f"Valor da função: {y}")
y.backward()
print(f"Derivada em x=2: {x.grad}")
print(f"Verificação manual: 3*2*2 + 2 = {3*2*2 + 2}")

# 2. Derivadas parciais de f(u,v) = u²v + uv²
print("\n=== Exercício 1B: Derivadas Parciais ===")
u = torch.tensor(3.0, requires_grad=True)
v = torch.tensor(2.0, requires_grad=True)
f = u**2 * v + u * v**2
print(f"Valor da função: {f}")
f.backward()
print(f"∂f/∂u = {u.grad}")
print(f"∂f/∂v = {v.grad}")

# 3. Visualização de função e derivada
print("\n=== Exercício 1C: Visualização ===")
x = torch.linspace(-5, 5, 100, requires_grad=True)
y_func = x**3 - 3*x**2 + 2*x
y_scalar = torch.sum(y_func)
y_scalar.backward()

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.plot(x.detach().numpy(), y_func.detach().numpy())
plt.title('Função y = x³ - 3x² + 2x')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x.detach().numpy(), x.grad.detach().numpy())
plt.title('Derivada dy/dx')
plt.grid(True)
plt.tight_layout()
plt.show()
```
